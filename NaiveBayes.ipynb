```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
```

```python
dataframe=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv')
test_dataframe=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv')
passengerId=test_dataframe["PassengerId"]
```

```python
final_dataframe=dataframe[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
final_dataframe=final_dataframe.dropna()
final_dataframe.head()
```

```python
final_dataframe['Sex']=final_dataframe["Sex"].replace(to_replace=final_dataframe['Sex'].unique(),value=[1,0])
```

```python
final_dataframe=pd.get_dummies(final_dataframe,drop_first=True)
final_dataframe
```

```python
train_y=final_dataframe['Survived']
train_x=final_dataframe[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked_Q','Embarked_S']]
```

```python
from sklearn.model_selection import train_test_split
train_data,val_data,train_target,val_target=train_test_split(train_x,train_y,train_size=0.8)
```

```python
model=GaussianNB()
model.fit(train_data,train_target)
val_pred=model.predict(val_data)
```

```python
from sklearn.metrics import accuracy_score
print('Model accuracy Score:(0:0.4f)'.format(accuracy_score(val_target,val_pred)*100)+'%')
```
