```python
data['buying'],_=pd.factorize(data['buying'])# _ is used to avoid separate variable creation
data['maint'],_=pd.factorize(data['maint'])
data['doors'],_=pd.factorize(data['doors'])
data['persons'],_=pd.factorize(data['persons'])
data['lug_boot'],_=pd.factorize(data['lug_boot'])
data['safety'],_=pd.factorize(data['safety'])
data.head()
```

```python
x=data.iloc[:,0:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0) #random_state=0 means that the values are taken in order if it is given some value it ranodmizes accoridngly ususally 42 is used
```

```python
x.columns
```

```python
from sklearn import tree,metrics
dtree=tree.DecisionTreeClassifier(criterion='gini')#entropy or gini
dtree.fit(x_train,y_train)
```

```python
y_train_pred=dtree.predict(x_train)
```

```python
accuracy=metrics.accuracy_score(y_train,y_train_pred)
print('Train accuracy: {:.2f}'.format(accuracy))
```

```python
y_test_pred=dtree.predict(x_test)
```

```python
accuracy=metrics.accuracy_score(y_test,y_test_pred)
print('Train accuracy: {:.2f}'.format(accuracy))
```
