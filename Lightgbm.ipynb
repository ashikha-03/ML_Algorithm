```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarning("ignore")
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Breast_cancer_data.csv')
```

```python
df
```

```python
x=df[['mean_radius','mean_texture','mean_perimeter','mean_area','mean_smoothness']]
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
import lightgbm as lgb
clf=lgb.LGBMClassifier()
clf.fit(x_train,y_train)
```

```python

y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_pred,y_pred)
print("Accuracy:{0:0.4f}".format(accuracy_score(y_test,y_pred)))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={
    'learning_rate':[0.25,0.5,0.75],
    'max_depth':[3,4,5],
    'subsample':[0.8,0.9,1.0],
    'colsample_bytree':[0.8,0.9,1.0],
    'n_estimators':[300,400,500]
    }
grid_search=GridSearchCV(estimator=clf,param_grid=param_grid,scoring='accuracy',cv=5,verbose=1)
```

```python
grid_search.fit(x_train,y_train)
print("best parameters",grid_search.best_params_)
print("best score",grid_search.best_score_)
```
